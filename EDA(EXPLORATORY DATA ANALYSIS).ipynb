# Exploratory Data Analysis(EDA)

# Value_counts

```python
# creating a function named value_counts and using a loop to show the count of values (0,1) in symptoms columns
def value_counts():
    for col in df.columns:
        print(col)
        print("="*30)
        display(df[col].astype(str).value_counts())
        print("\n\n")
        
value_counts()
```

# Univariate Analysis

```python
#creating a function called analysis for visualization of symptoms occuring in diseases through barplot
def analysis():
    for col in df.columns:
        print(col)
        if col!= "prognosis":
            group_data = df.groupby("prognosis").agg(count_1 = (col,"sum"),
                                                    total_count = (col,"count")).reset_index()
            if group_data["count_1"].sum() == 0:
                continue
            else:
                sns.barplot( data= group_data[group_data["count_1"] > 0],
                x = "prognosis",
                y = "count_1")
                plt.xticks(rotation=90)
                plt.show()
```

```python
analysis()
```

# Bivariate Analysis

```python
group_data = df.groupby("prognosis").sum()
```

```python
df_tidy = pd.melt(
    df.groupby("prognosis").sum().reset_index(),
    id_vars=["prognosis"],
    var_name = "Symptom",
    value_name = "Result"
)
df_tidy.head()
```

```python
# creating a pivot table to show the count of symptoms in the diseases
df_tidy_2 = pd.pivot(
    df_tidy,
    index = "Symptom",
    columns = "prognosis",
    values = "Result"
).reset_index()

df_tidy_2.columns.name = None
df_tidy_2.head()
```

```python
# creating function called bi_analysis for visualization of bivariate analysis between symptoms and prognosis through
# the pivot table formed above as df_tidy_2

def bi_analysis():
    for col in df_tidy_2.columns:
        print(col)
        if col not in ["prognosis","Symptom"]:
            if df_tidy_2[col].sum() == 0:
                continue
            else:
                sns.barplot( 
                    data= df_tidy_2[df_tidy_2[col] > 0],
                    x = "Symptom",
                    y = col
                )
                plt.xticks(rotation=90)
                plt.show()
```

```python
bi_analysis()
```
