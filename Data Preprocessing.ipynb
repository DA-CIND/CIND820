# Outlier

```python
fig, ax = plt.subplots(ncols=5, nrows=2, figsize=(30,10))
index = 0
ax = ax.flatten()

for attributeCol, value in df.iloc[:, :10].items():
    sns.boxplot(y=attributeCol, data=df.iloc[:, :10], ax=ax[index], color="#cf7849")
    index += 1
plt.tight_layout(pad=1.5, w_pad=1.5, h_pad=5.0)

plt.suptitle('Disease Prediction Box Plot',y=1.01, size = 16, color = 'black', weight='bold')
```

From the above subplots it can be infered that there are no outliers in the dataset. As the dataset has binary values.

# Balancing the Dataset

```python
# Checking whether the dataset is balanced or not
disease_counts = df["prognosis"].value_counts()
temp_df = pd.DataFrame({
    "Disease": disease_counts.index,
    "Counts": disease_counts.values
})
 
plt.figure(figsize = (20,10))
sns.barplot(x = "Disease", y = "Counts", data = temp_df)
plt.xticks(rotation=90)
plt.show()
```

The above graph shows that the dataset is balanced and there is no need for furture balancing.
